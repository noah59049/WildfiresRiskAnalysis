```python
def hist6(colname):
    plt.hist(NRI6[colname])
    if colname.startswith("Wildfire - "):
        colname = colname[11:]
    plt.xlabel(colname)
    plt.gca().xaxis.set_major_formatter(ticker.FuncFormatter(lambda x, _: f'{x:.0f}'))  # Format x-axis as integers

    # plt.show()
```

```python
plt.figure(figsize=(15, 15))
i = 0
for colname in NRI6.columns:
    if colname.startswith("Wildfire"):
        i += 1
        plt.subplot(5,5, i)
        hist6(colname)
        plt.tight_layout(rect=[0, 0.03, 1, 0.95])
```

```python
# I wonder if removing zeros will help with wacky log plots
def hist6a(colname):
    column = NRI6[colname]
    column = column[column > 0]
    plt.hist(np.log10(column))
    # plt.xscale("log")
    if colname.startswith("Wildfire - "):
        colname = colname[11:]
    elif colname.startswith("Expected Annual Loss"):
        colname = colname[20:]
    plt.xlabel(colname)
    plt.gca().xaxis.set_major_formatter(ticker.FuncFormatter(lambda x, _: f'{x:.0f}'))  # Format x-axis as integers

    # plt.show()
```
